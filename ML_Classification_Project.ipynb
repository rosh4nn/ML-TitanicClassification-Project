```python
x = df.drop(['Name','Survived','Ticket','Embarked'],axis=1)
x
```

```python
y = df['Survived']
y
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()
x_train,y_train = smote.fit_resample(x_train,y_train)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train1 = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

knn = KNeighborsClassifier()
knn
```

```python
knn.fit(x_train1,y_train)
```

```python
y_pred = knn.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report

print(confusion_matrix(y_test,y_pred))
knn_acc = accuracy_score(y_test,y_pred)
print(knn_acc)
print(classification_report(y_test,y_pred))
```

HyperParameter Tuning

```python
knn1 = KNeighborsClassifier()
parameters = {'n_neighbors':[1,3,5,7,9],
              'weights':['uniform', 'distance'],
              'algorithm':['auto', 'ball_tree', 'kd_tree', 'brute'],

              }
```

```python
gs = GridSearchCV(knn,parameters,cv=10,scoring='accuracy')
gs.fit(x_train1,y_train)
```

```python
gs.best_params_
```

```python
knn1 = KNeighborsClassifier(algorithm='auto',n_neighbors=7,weights='distance')
knn1.fit(x_train1,y_train)
```

```python
y_pred_knn1 = knn1.predict(x_test)
y_pred_knn1
```

```python
print(confusion_matrix(y_test,y_pred_knn1))
grid_knn = accuracy_score(y_test,y_pred_knn1)
print(grid_knn)
print(classification_report(y_test,y_pred_knn1))
```

```python
import matplotlib.pyplot as plt

models = [
    "KNN",
    "KNN Grid",
    "SVM",
    "SVM Grid",
    "Decision Tree",
    "Decision Tree Grid",
    "Logistic Regression",
    "Random Forest",
    "Random Forest Grid"]

accuracy = [knn_acc,grid_knn,svm_acc,grid_svm,tree_acc,grid_tree,logistic_acc,rf_acc,grid_rf]

plt.figure(figsize=(8,5))

plt.bar(models, accuracy)

plt.xticks(rotation=45)
plt.xlabel("Models")
plt.ylabel("Accuracy")
plt.title("Model Accuracy Comparison")

plt.ylim(0, 1)

plt.show()
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

y_score = rf1.predict_proba(x_test)[:,1]

fpr, tpr, _ = roc_curve(y_test, y_score)

auc = roc_auc_score(y_test, y_score)

plt.plot(fpr, tpr, label=f"AUC = {auc:.2f}")
plt.plot([0,1], [0,1], "--")

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()

plt.show()

print("ROC AUC Score:", auc)
```
